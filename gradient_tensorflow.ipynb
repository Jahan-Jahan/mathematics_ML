```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
X = tf.constant([[1.0], [2.0], [3.0], [4.0]])
y = tf.constant([[3.1], [5.68], [6.38], [9.34]])
```

```python
plt.plot(X, y, "or")
plt.show()
```

```python
w = tf.Variable(tf.random.normal([1]))
b = tf.Variable(tf.random.normal([1]))
```

```python
learning_rate = 0.01
epochs = 1000
```

```python
for i in range(epochs):
    with tf.GradientTape() as tape:
        y_pred = X * w + b
        loss = tf.reduce_mean(tf.square(y - y_pred))

    gradient = tape.gradient(loss, [w, b])
    w.assign_sub(learning_rate * gradient[0])
    b.assign_sub(learning_rate * gradient[1])

    print(f"Epoch {i + 1}, Loss: {loss.numpy()}, Weight: {w.numpy()[0]}, Bias: {b.numpy()[0]}")

print(f"Linear Equation: ({w.numpy()[0]})X + ({b.numpy()[0]})")
```

```python
plt.plot(X, y, "or")
plt.plot(X, w * X + b, "-b")
plt.show()
```
