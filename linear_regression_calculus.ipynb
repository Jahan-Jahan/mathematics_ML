```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1], [3], [5], [7]])
X
```

```python
y = np.array([5, 7, 10, 14])
y
```

```python
w = np.random.randn(1) # Initializing random weight
b = np.random.randn() # Initializing random bias
print(w, b)
```

```python
learning_rate = 0.01
epochs = 1000
n = len(y)
for i in range(epochs):
    y_pred = X @ w + b
    loss = np.mean((y_pred - y) ** 2)
    dw = -2/n * X.T @ (y - y_pred)
    db = -2/n * np.sum(y - y_pred)
    w -= learning_rate * dw
    b -= learning_rate * db
    print(f"Epoch: {i + 1} , Loss: {loss}")

print(f"Final Weights: {w}")
print(f"Final Bias: {b}")

y_final_pred = X @ w + b
print(f"Final Prediction: {y_final_pred}")
```
